```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'class_all_final_test.csv', encoding='utf-8')
print (df)
```

```python
print(df.keyword)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 

tfidf_vectorizer = TfidfVectorizer() # TF-IDF 객체선언
tfidf_matrix = tfidf_vectorizer.fit_transform(df['keyword'])
```

```python
print(tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
np.round(cosine_matrix, 4)
```

```python
class2id = {}
for i, c in enumerate(df['class']): class2id[i] = c

id2class = {}
for i, c in class2id.items(): id2class[c] = i
```

```python
class2id
```

```python
id2class
```

```python
idx = id2class['검색']
sim_scores = [(i, c) for i, c in enumerate(cosine_matrix[idx]) if i != idx] # 자기 자신을 제외한 영화들의 유사도 및 인덱스를 추출 
sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse=True) # 유사도가 높은 순서대로 정렬 
sim_scores[0:10] # 상위 10개의 인덱스와 유사도를 추출 
```

```python
sim_scores = [(class2id[i], score) for i, score in sim_scores[0:10]]
sim_scores
```
