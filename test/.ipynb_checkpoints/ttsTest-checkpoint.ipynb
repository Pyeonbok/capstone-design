## text 추출 단계 성공

```python
raw = parser.from_file('sample.pdf')

list=[]
list.append(raw['content'])
text = " ".join(list)

num=0
list2=[]
temp=""

for i in range(len(text)):
    if text[i]=="\n":
        print(temp)
        list2.append(temp)
        temp=""
        num+=1
    else : 
        temp+=text[i]

print(len(list2))


#print(raw['content'])
```
