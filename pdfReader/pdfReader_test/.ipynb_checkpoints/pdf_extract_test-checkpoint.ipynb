```python
from PyPDF2 import PdfFileReader, PdfFileWriter
from tika import parser
import tabula
```

### 1. PDF 첫 페이지 제거

```python
pdf = PdfFileReader(open('sample2.pdf','rb'))

# pdf페이지수
numberPages = pdf.getNumPages()

# 빈 pdf를 하나 만들기
pdf_writer = PdfFileWriter()
for page in range(1, numberPages):
    
    # 두번째 페이지부터 마지막 페이지까지 빈 pdf에 추가하기
    pdf_writer.addPage(pdf.getPage(page))
    output_name = 'sample_extract_{}.pdf'.format('2')
#    save_path = os.path.join(path, output_name)

    # 파일 쓰기하면서 저장
    with open(output_name, 'wb') as f:
        pdf_writer.write(f)
```
